```python
import pandas as pd
```

```python
data = pd.read_csv('building_v1.csv')
```

```python
count = data['X'].count()
```

```python
cleaned_data = data.groupby(['ISSUED_YEAR','B1_PER_TYPE']).count().drop(columns=['Y']).rename(columns={'X':'count'}).reset_index()
```

```python
cleaned_data
```

```python
cleaned_data['precentage'] = cleaned_data['count']/count
```

```python
cleaned_data.to_csv('buildings_group.csv',index=False)
```
